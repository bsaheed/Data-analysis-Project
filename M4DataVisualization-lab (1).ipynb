## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
query ='''
SELECT ConvertedComp from master'''

Converted = pd.read_sql_query(query, conn)

Converted.plot.hist(bins= 12)
plt.title('Histogram of ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
query= ''' SELECT Age from master'''

age = pd.read_sql_query(query, conn)
sns.boxplot(age, orient ='h')
plt.title('Boxplot of Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query ='''select Age, WorkWeekHrs from master'''

data = pd.read_sql_query(query, conn)

sns.scatterplot(y= data['Age'], x =data['WorkWeekHrs'])
plt.title('Scatter plot of Age and Hours worked in week')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

query = ''' select Age, WorkWeekHrs, CodeRevHrs from master'''

data = pd.read_sql_query(query, conn)
#data.isnull().sum()
#data.dropna(subset ='Age', axis=0, inplace =True )
sns.scatterplot(data=data, x= 'WorkWeekHrs', y ='CodeRevHrs',size ='Age')
#px.scatter(data_frame=data, x ='CodeRevHrs', y= 'WorkWeekHrs', size = 'Age', title= 'Bubble plot of WorkWeekHrs against CodeRevHrs ')

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query = ''' SELECT DatabaseDesireNextYear, Count(*)as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by Count desc limit 5'''

data = pd.read_sql_query(query, conn)
data.head()
data.set_index('DatabaseDesireNextYear',inplace=True)

sizes=data.iloc[:,0]
color_list=['red','yellowgreen','gold','blue','lightcoral','pink']
labels=['PostgreSQL','MongoDB','Redis','MySQL','Elasticsearch']
data.plot(kind='pie',
               figsize=(20,6),
               autopct='%1.1f%%',
               startangle=90,
               shadow=True,
               labels=None,
               pctdistance=1.12,
               subplots=True,
               colors=color_list)

plt.title('Pie Chart of Top 5 Database Desire Next Year')
plt.axis('equal')
plt.legend(labels, loc='upper left')
plt.show()
#data.plot(kind ='pie')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query= ''' SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35

'''
data = pd.read_sql_query(query, conn)
data.head()
df =data.groupby('Age')['WorkWeekHrs', 'CodeRevHrs'].median()
df.plot(kind='bar', stacked= True, title ='A stacked chart of the median of WorkWeekHrs and CodeRevHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
query = ''' SELECT Age, ConvertedComp from master WHERE Age BETWEEN 25 AND 30'''

data = pd.read_sql_query(query, conn)
df= data.groupby('Age').median()
df.plot(kind= 'line', title ='Line Chart of ConvertedComp by Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
query =''' SELECT MainBranch, Count(MainBranch) As Count FROM master GROUP BY MainBranch '''
data = pd.read_sql_query(query, conn)
data.head()
data.set_index('MainBranch', inplace= True)
data.plot(kind ='barh', color ='pink')
plt.title('Bar Chart of Main Branch')
plt.show()
```
